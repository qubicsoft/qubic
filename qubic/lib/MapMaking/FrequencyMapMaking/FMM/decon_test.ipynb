```python
import pickle
import numpy as np
import healpy as hp
from astropy.coordinates import SkyCoord
import astropy.units as u

import matplotlib.pyplot as plt
```

```python
plt.rcParams["text.usetex"] = True
```

```python
moon_decon = "/home/sofiafera/thesis/qubic/qubic/lib/MapMaking/FrequencyMapMaking/FMM/test_master_moon_100_deconv/maps/test_None.pkl"
moon_no_decon = "/home/sofiafera/thesis/qubic/qubic/lib/MapMaking/FrequencyMapMaking/FMM/test_master_moon_100no_deconv/maps/test_None.pkl"
```

```python
def read_pickle(path: str):
    with open(path, "rb") as fin:
        return pickle.load(fin)
```

```python
decon = read_pickle(moon_decon)
no_decon = read_pickle(moon_no_decon)
```

```python
q = len(decon['maps_in'])
delta = decon["maps"] - no_decon["maps"]


# Plot all delta maps in a single figure with 2 rows (bands) and 3 columns (I,Q,U)
# Center the patch at RA=0, DEC=-57 (equatorial coordinates)
# Coordinate equatoriali
coord_eq = SkyCoord(ra=0*u.deg, dec=-57*u.deg, frame='icrs')
coord_gal = coord_eq.galactic


stokes = ["I", "Q", "U"]
rows = q   # number of bands to show as rows (0 and 1)
cols = 3   # I, Q, U as columns


fig = plt.figure(figsize=(4 * cols, 4 * rows))
for i in range(min(q, rows)):  # protect in case q > 2
    for s, st in enumerate(stokes):
        # panel index increases row-wise
        sub_index = i * cols + s + 1
        hp.gnomview(np.log10(delta[i, :, s] / delta[i, :, s].max()),
                    rot=[coord_gal.l.value, coord_gal.b.value],
                    reso=15,
                    xsize=200,
                    ysize=200,
                    unit="dB",
                    sub=(rows, cols, sub_index),
                    title=rf"$ \Delta \, {st} \, - band \ {i}$",
                    notext=False,
                    # min=-40,
                    max=0)

plt.savefig("delta_CR_minus_noCR_all.pdf", bbox_inches="tight", dpi=1000)
plt.close(fig)
```
