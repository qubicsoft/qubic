```python
all_datasets_23 = []
for file in pkl_23:
    data = utils.load_results(file)
    all_datasets_23.append(data)
```

```python
all_datasets_25 = []
for file in pkl_25:
    data = utils.load_results(file)
    all_datasets_25.append(data)
```

```python
all_amplitudes_23 = []

for data in all_datasets_23:
    offsets = data['results']['offset_dt']
    
    for tes in offsets:
        all_amplitudes_23.extend(offsets[tes])
```

```python
all_amplitudes_25 = []

for data in all_datasets_25:
    offsets = offsets = data['results'].get('offset_dt', {})
    
    for tes in offsets:
        all_amplitudes_25.extend(offsets[tes])
```

```python
all_amp_23 = np.array(all_amplitudes_23)
all_amp_25 = np.array(all_amplitudes_25)
```

```python
from collections import Counter

tes_dataset_count_23 = Counter()

for data in all_datasets_23:
    tes_keys = data['results']['offset_dt'].keys()
    tes_dataset_count_23.update(tes_keys)
```

```python
tes_dataset_count_25 = Counter()


for data in all_datasets_25:
    offsets = data['results'].get('offset_dt', {})  # devuelve {} si no existe
    tes_dataset_count_25.update(offsets.keys())
```

```python
tes_23 = list(tes_dataset_count_23.keys())
counts_23 = list(tes_dataset_count_23.values())
```

```python
tes_25 = list(tes_dataset_count_25.keys())
counts_25 = list(tes_dataset_count_25.values())
```

```python
plt.bar(tes_25, counts_25)
plt.xlabel("TES")
plt.ylabel("Number of datasets")
plt.title("TES occurrence across datasets 2025")
plt.show()
```

```python
plt.bar(tes_23, counts_23)
plt.xlabel("TES")
plt.ylabel("Number of datasets")
plt.title("TES occurrence across datasets 2023")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,5))

ax[0].plot(abs(all_amp_23), marker='s', color='royalblue', lw=0)

ax[0].axhspan(ymin=3.5e5, ymax=5.5e5, color='grey', alpha=0.3, label='Possible region')
ax[0].axhline(y=4.5e5, color='crimson', ls='--')
ax[0].axhline(y=9e5, color='black')
#ax[0].axhline(y=2.25e5, color='black')
ax[0].set_ylim(2e5, 1.7e6)
ax[0].set_ylabel('Amplitude [ADUs]', fontsize=15)
ax[0].set_xlabel('Number of flux jumps', fontsize=15)
ax[0].set_title('Datasets year 2023', fontsize=15)
ax[0].legend(fontsize=15)

ax[1].hist(abs(all_amp_23), bins=30, range=(3e5, 1e6), color='royalblue', histtype='barstacked')#, edgecolor='black')
ax[1].set_ylabel('Number of flux jumps', fontsize=15)
ax[1].set_xlabel('Amplitude [ADUs]', fontsize=15)
ax[1].set_title('Datasets year 2023', fontsize=15)
ax[1].axvline(x = 0.49e6, color='crimson', ls='--')
ax[1].axvline(x = 0.46e6, color='crimson', ls='--')

plt.savefig('statistics2_23.pdf', bbox_inches='tight', pad_inches=0)
```
