```python
from qubic.lib.MapMaking.Qmap_plotter import plot_cross_spectrum
from qubic.lib.Qhdf5 import HDF5Dict
```

```python
data = HDF5Dict().load_dict("parametric_d0_DB_test_cmm/Spectrum/spectrum_test_None.data")
data.keys()
```

```python
ell = data["ell"]
Dls = data["Dls"]
Nls = data["Nls"]

Dls_noiseless = Dls - Nls

print("ell", ell.shape, ell)
print("Dls", Dls.shape)
print("Nls", Nls.shape)
```

```python
N = Dls.shape[0]
comp = ["CMB", "Dust"]
plot_cross_spectrum(
    nus=comp,
    ell=ell,
    Dl=Dls,
    Dl_err=Nls,
    ymodel=None,
    mode="Dl",
    nrec=N,
    figsize=(12, 10),
    title="",
)
```
