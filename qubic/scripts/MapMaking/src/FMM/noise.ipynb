```python
import healpy as hp
import scipy.constants as c

def NEP2NET(nep, delta_nu, eta):
    return nep/(2*c.k*delta_nu*eta)
  
    
def sigma_pixel(instr, dic, noise_eq_unit, fsky):
    Ndet = len(instr)
    npix = fsky*hp.nside2npix(nside=dic['nside'])
    
    N_hits = dic['npointings']*Ndet/npix

    sigma_tod = noise_eq_unit/np.sqrt(2*dic['period']) # W/sqrt(Hz) or T/sqrt(Hz)

    sigma_pixel_theo = sigma_tod/np.sqrt(N_hits)
    
    return sigma_pixel_theo # W or T
    

def sigma_pixel_per_sqrt_tobs(noise_eq_unit, fsky):
    
    Ndet = 256 #len(instr)    
    theta_pixel = hp.nside2resol(nside=64) # rad , arcmin=True)/60.

    sigma_pixel_per_sqrt_tobs = noise_eq_unit*np.sqrt(4*c.pi*fsky)/np.sqrt(2*Ndet)/theta_pixel
    return sigma_pixel_per_sqrt_tobs # W*sqrt(s) or T*sqrt(s)

def sigma_pixel_given_tobs(noise_eq_unit, fsky, tobs):
    return sigma_pixel_per_sqrt_tobs(noise_eq_unit, fsky)/np.sqrt(tobs)
```

## Photon noise NEP

$$NEP^2 = 2h^2\nu^2\Delta\nu^2\left(\frac{n}{\eta}+n^2\right)$$

```python
import numpy as np
h = 6.626e-34
nu = 150e9
delta_nu = 40e9
k_b = 1.38e-23
T = 15
eta = 1

n = 1 / (np.exp(h*nu/(k_b*T))-1)

nep = np.sqrt( 2*(h*nu)**2 * delta_nu * (n/eta+n**2) )
nep
```
