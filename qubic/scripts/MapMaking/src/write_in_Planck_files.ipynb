```python
import healpy as hp
import numpy as np
import pickle
import os

nside = 256
npix = 12 * nside**2
```

```python

planck_info = {
    "frequency": [30, 44, 70, 100, 143, 217, 353],
    "depth_i": [150.0, 162.0, 210.0, 77.4, 33.0, 46.8, 154],
    "depth_p": [210.0, 240.0, 300.0, 118, 70.2, 105.0, 439],
}

base_path = "/home/lkardum/qubic/qubic/data"
```

```python

for i, nu in enumerate(planck_info["frequency"]):
    filename = os.path.join(base_path, f"Planck{nu}GHz.pkl")

    print(f"Updating: {filename}")

    with open(filename, "rb") as f:
        planck_data = pickle.load(f)

    # calculate sigmas 
    sigma_I = planck_info["depth_i"][i] / hp.nside2resol(nside, arcmin=True)
    sigma_P = planck_info["depth_p"][i] / hp.nside2resol(nside, arcmin=True)
    sigma = [sigma_I, sigma_P, sigma_P]
    #print(sigma)

    # one noise realization (not really neccessary) 
    noise = np.random.standard_normal((npix, 3)) * sigma
    #print(noise.shape)
    
    planck_data[f"sigma{nu}"] = sigma
    planck_data[f"noise{nu}"] = noise

    with open(filename, "wb") as f:
        pickle.dump(planck_data, f)

print("updated all")
```
